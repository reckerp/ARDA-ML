```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dog = pd.read_csv('../Data/ARDA ML - Dog.csv') # 13000 iteration
df_wheel = pd.read_csv('../Data/ARDA ML - Wheel.csv') # 14000 iteration
df_dog.columns = df_dog.columns.str.strip()
df_wheel.columns = df_wheel.columns.str.strip()
df_dog.rename(columns={'Picture 1': '1', 'Picture 2': '2', 'Picture 3': '3'}, inplace=True)
df_wheel.rename(columns={'Picture 1': '1', 'Picture 2': '2', 'Picture 3': '3'}, inplace=True)
```

```python
# Dog
row_dog = df_dog[df_dog['Iteration'] == 13000]

# Plot
row_dog.iloc[0, 1:].plot(kind='bar')
plt.xticks(rotation=0)

plt.ylabel('Accuracy (%)')
plt.xlabel('Picture (Dog)')

export_path = '../Data/dog-outliers.png'
plt.savefig(export_path, dpi=300)

plt.show()
```

```python
# Wheel
row_wheel = df_wheel[df_wheel['Iteration'] == 14000]

# Plot
row_wheel.iloc[0, 1:].plot(kind='bar')
plt.xticks(rotation=0)

plt.ylabel('Accuracy (%)')
plt.xlabel('Picture (Wheel)')

export_path = '../Data/wheel-outliers.png'
plt.savefig(export_path, dpi=300)

plt.show()
```
