```python
import json
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv('../Data/ARDA ML - Dog.csv')

df.columns = df.columns.str.strip()

# Plotting 'Iteration' against 'Picture 1' after trimming any potential whitespace
plt.figure(figsize=(10, 6))
plt.plot(df['Iteration'], df['Picture 1'], marker='o', linestyle='-', color='#FF6961')

# Adding titles and labels
plt.xlabel('Training Iterations')
plt.ylabel('Accuracy (%)')
plt.grid(True)

output_file_path = '../Data/dogs_image1_accuracy_vs_iteration.png'
plt.savefig(output_file_path, facecolor='white', bbox_inches='tight')

#plt.savefig(output_file_path)

plt.show()
```
