```python
import pandas as pd

import matplotlib.pyplot as plt


df = pd.read_csv('../Data/ARDA ML - Wheel.csv')

df.columns = df.columns.str.strip()

plt.figure(figsize=(10, 6))
plt.plot(df['Iteration'], df['Picture 3'], marker='o', linestyle='-', color='#FF6961')

plt.xlabel('Training Iterations')
plt.ylabel('Accuracy (%)')
plt.grid(True)

output_file_path = '../Data/wheel_image1_accuracy_vs_iteration.png'
plt.savefig(output_file_path, facecolor='white', bbox_inches='tight')

plt.show()
```
