```python
import json
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/ARDA ML - Loss (1).csv')

df.columns = df.columns.str.strip()

coefficients = np.polyfit(df['Iteration'], df['loss'], 2)

polynomial_model = np.poly1d(coefficients)

x_values = np.linspace(0, df['Iteration'].max(), 500)  # More points for a smoother line

fitted_values = polynomial_model(x_values)

plt.figure(figsize=(10, 6))
plt.scatter(df['Iteration'], df['loss'], label='Data Points', color='#1b5886', alpha=0.6)
plt.plot(x_values, fitted_values, 'r-', label='Quadratic Trend Line', color='#FF6961')

plt.xlim(0, 21000)


plt.xticks(np.arange(0, 21001, 2000))


plt.yticks(np.arange(4, 13, 1))

plt.xlabel('Number of training iteration', fontsize=12)
plt.ylabel('Loss', fontsize=12)

plt.legend()
plt.grid(True)

output_file_path = '../Data/loss_by_iteration_plot.png'

plt.savefig(output_file_path)

plt.show()
```
