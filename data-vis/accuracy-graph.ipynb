```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../Data/accuracy-averages-classes.csv")
```

```python
for column in df.columns[1:]:
    plt.plot(df['Iteration'], df[column], label=column, linewidth=2.5)

fig = plt.gcf()  # Get the current figure
fig.set_size_inches(12, 6)  # Set the size of the figure

plt.xlabel('Iteration')
plt.ylabel('Accuracy (%)')
plt.title('')
plt.legend(loc='lower right')  # Place the legend at the bottom right
plt.grid()

# Customize the tick marks on the x-axis to range from 0 to 20000 with steps of 2000
plt.xticks(np.arange(0, 21001, 2000))

export_path = '../Data/accuracy-graph.png'
plt.savefig(export_path, dpi=300)

plt.show()
```
